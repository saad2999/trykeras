```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
TrainLables=[]
TrainSmps=[]
```

```python
for i in range(50):
  randomY=randint(13,64)
  TrainSmps.append(randomY)
  TrainLables.append(1)


  randomO=randint(65,100)
  TrainSmps.append(randomO)
  TrainLables.append(0)

  for i in range(1000):
    randomY=randint(13,64)
    TrainSmps.append(randomY)
    TrainLables.append(0)


  randomO=randint(65,100)
  TrainSmps.append(randomO)
  TrainLables.append(1)
```

```python
print(TrainSmps)
```

```python
TrainLables=np.array(TrainLables)
TrainSmps=np.array(TrainSmps)
TrainLables,TrainSmps=shuffle(TrainLables,TrainSmps)
print((TrainSmps))
```

```python
sclars=MinMaxScaler(feature_range=(0,1))
sclars_t_smap=sclars.fit_transform(TrainSmps.reshape(-1,1))
for i in sclars_t_smap:
  print(i)
```
